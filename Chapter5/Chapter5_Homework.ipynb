1. Plot the rolling 20-day minimum of the Facebook closing price using pandas.

```python
%matplotlib inline
import pandas as pd


fb = pd.read_csv('./data/fb_stock_prices_2018.csv', index_col='date', parse_dates=True)

fb.close.rolling('20D').min().plot(title='Rolling 20-day Minimum of the Facebook stock closing price')

```

2. Create a histogram and KDE of the change from open to close in the price of Facebook stock.

```python
diff = fb.open - fb.close
histo = diff.plot(kind='hist', density=True, alpha=0.3)
diff.plot(kind='kde', ax=histo, title='NASDAQ: FB Daily Open/Close Price Difference')
```

4. Make a line plot of the difference between the weekly maximum high price and the weekly minimum low price for Facebook. This should be a single line.

```python
wdf = fb.resample('1W').agg(dict(high='max', low='min'))
wdf = wdf.assign(max_weekly_diff = lambda x: x.high - x.low)

wdf.max_weekly_diff.plot(title = 'NASDAQ: FB: Difference between the Highest and the Lowest Wekkly Price.')
```

5. Plot the 14-day moving average of the daily change in new COVID-19 cases in Brazil, China, India, Italy, Spain, and the USA:

a) First, use the `diff()` method that was introduced in the Working with time series data section of Chapter 4, Aggregating Pandas DataFrames, to calculate the day-over-day change in new cases. Then, use `rolling()` to calculate the 14-day moving average.

```python
import matplotlib.pyplot as plt

covid = pd.read_csv('./data/covid19_cases.csv').assign(
    date=lambda x: pd.to_datetime(x.dateRep, format='%d/%m/%Y')
).set_index('date')

covid = covid.replace(
    'United_States_of_America', 'USA'
).sort_index()

rolling_average_new_caes = covid.pivot_table(index =  covid.index, columns=['countriesAndTerritories'], values='cases')
rolling_average_new_caes.apply(lambda x: x.diff().rolling(14).mean())
rolling_average_new_caes.tail()
```

b) Make three subplots: one for China; one for Spain and Italy; and one for Brazil, India, and the USA.

```python
fig, ax = plt.subplots(1, 3, figsize=(21, 7))
rolling_average_new_caes[['China']].plot(ax=ax[0], 
                        title='14-day rolling average of change in daily new COVID-19 cases\n in China')
rolling_average_new_caes[['Italy', 'Spain']].plot(ax=ax[1],
                        title='14-day rolling average of change in daily new COVID-19 cases\n in Italy & Spain')
rolling_average_new_caes[['Brazil', 'USA', 'India']].plot(ax=ax[2], 
                        title='14-day rolling average of change in daily new COVID-19 cases\n in Brazil, USA & India')
```

6. Using matplotlib and pandas, create two subplots side-by-side showing the effect that after-hours trading has had on Facebook's stock prices:

    a) The first subplot will contain a line plot of the daily difference between that day's opening price and the prior day's closing price.

```python
fig, ax = plt.subplots(1, 2, figsize=(21, 7))

fb = fb.assign(prior_day_close=lambda x: x.close.shift(),
         change_in_price = lambda x: x.open - x.prior_day_close)

fb.change_in_price.plot(ax=ax[0]
                       , title = 'After Hours Trading \n (Current Day\'s Ope Price - Prior Day\' Closing Price)')
                       
monthly_net_effect = fb.change_in_price.resample('1M').sum()
monthly_net_effect.index = pd.to_datetime(monthly_net_effect.index, format = '%m/%d/%Y')
monthly_net_effect.index = monthly_net_effect.index.strftime('%Y-%m-%d')

monthly_net_effect.plot(ax=ax[1]
                       , kind='bar'
                       , title='Monthly Effect of After Hours Trading')

monthly_net_effect.head()

#(fb.open - fb.close.shift())
```

    c) Bonus #1: Color the bars according to whether there are gains in the stock price (green) or drops in the stock price (red).

```python
import numpy as np
monthly_net_effect.plot(kind='bar',
                title='Monthly Effect of After Hours Trading'
                , color=np.where(monthly_net_effect >= 0, 'g', 'r'))
```

    d) Bonus #2: Modify the x-axis of the bar plot to show the three-letter abbreviation for the month.

```python

monthly_net_effect.index = pd.to_datetime(monthly_net_effect.index)
monthly_net_effect.index = monthly_net_effect.index.strftime('%b')

monthly_net_effect.plot(kind='bar'
                       , title='Monthly Effect of After Hours Trading'
                       , color=np.where(monthly_net_effect >= 0, 'g', 'r'))
```
